```python
import cv2
import os
from tqdm import tqdm

def extract_frames(video_path, output_folder):
    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Open the video file
    video_capture = cv2.VideoCapture(video_path)

    # Get total number of frames in the video
    total_frames = int(video_capture.get(cv2.CAP_PROP_FRAME_COUNT))

    # Initialize tqdm progress bar
    progress_bar = tqdm(total=total_frames, desc='Extracting frames', unit='frame')

    # Read the video frame by frame
    success, image = video_capture.read()
    count = 0
    while success:
        # Save the extracted frame every 10th frame
        if count % 10 == 0:
            frame_filename = f"frame_{count}.jpg"
            frame_path = os.path.join(output_folder, frame_filename)
            cv2.imwrite(frame_path, image)

        # Update progress bar
        progress_bar.update(1)

        # Read next frame
        success, image = video_capture.read()
        count += 1

    # Release the video capture object
    video_capture.release()

    # Close tqdm progress bar
    progress_bar.close()

# Example usage
video_file = "DJI_20231125134955_0002_D.MP4"
output_folder = "extracted_frames"
extract_frames(video_file, output_folder)
```
